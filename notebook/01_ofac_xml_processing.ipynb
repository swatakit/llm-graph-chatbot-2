# Download XML and Convert to JSON format

- Let LLM do summrisation of the metadata

```python
# notebook/01_ofac_xml_processing.ipynb

from dotenv import load_dotenv
import os
import sys

sys.path.append(os.path.abspath('..')) 
load_dotenv('../.env',override=True)

from notebook.chain.ofac_analyzer import analyze_ofac_data
import json

def process_ofac_data():
    try:

        # Process OFAC data
        result =  analyze_ofac_data()

        # Print statistics and sample data
        if result and isinstance(result, dict):

            return result


    except Exception as e:
        print(f"Error during OFAC processing: {str(e)}")
        return None

# Execute the async function
try:
    result =  process_ofac_data()
    
    if result:
        # Save results to file (optional)
        output_file = "ofac_data.json"
        with open(output_file, 'w') as f:
            json.dump(result, f, indent=2)
        print(f"\nResults saved to {output_file}")
        print("\n=== Data Summary ===")
        print(f"Total Records: {result.get('total_records', 0)}")
        print(f"Individual Records: {len(result.get('individuals', []))}")
        print(f"Entity Records: {len(result.get('entities', []))}")
                
        print("\n== Overall Analysis ===\n")
        print(result['analysis'])
        
except Exception as e:
    print(f"Error running process: {str(e)}")
```
